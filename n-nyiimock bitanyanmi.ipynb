# importing the necessary library

```python
import PyPDF2
```

# Loading the pdf file

```python
file_p = open('assignment.pdf','rb')
```

# Directing the PyPDF2 to the file laoded

```python
reader = PyPDF2.PdfFileReader(file_p)
```

# Creating a function that will extract the pdf file for all the pages selected

```python
page_content = ""
for page_number in range(10):
    page48 = reader.getPage(48)
    page_content += page48.extractText()
    print(page_content)
```

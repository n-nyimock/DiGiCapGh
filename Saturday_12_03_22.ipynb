```python
mask=np.array(Image.open('ghanaflag.png'))
```

```python
fl=open('sampletext.txt')
```

```python
text=fl.read()
```

```python
wordcloud_gh=WordCloud(background_color='yellow',mode="RGB",max_words=1000,mask=mask).generate(text)
```

```python
image_colors=ImageColorGenerator(mask)
```

```python
plt.figure(figsize=[13,15])
plt.imshow(wordcloud_gh.recolor(color_func=image_colors),interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
avgs=grpby['price'].mean()
avgs
```

```python
averages={}
for make in df['make'].unique():
    tempdf=df[df['make']==make]
    average=tempdf['price'].mean()
    averages[make]=[average]
    
aggregate_df=pd.DataFrame.from_dict(averages,orient='index',columns=['AVERAGE PRICE'])
print(aggregate_df)
```

```python
aggs=df.groupby('make').agg(mean_credit_amount=pd.NamedAgg('price','mean'),
                           standard_deviation_credit_amount=pd.NamedAgg('price','std'),
                           variance_credit_amount=pd.NamedAgg('price','var'))
aggs
```

```python
def group_range(x):
    return x.max() - x.min()
ranges=df.groupby(['make','fuel_type'])['price'].apply(group_range)
print(ranges)
```

```python
range1=df.groupby(['make','fuel_type']).apply(lambda x: x['price'].max()<20000)
print(range1)
```

```python
range2=df.groupby(['make','fuel_type']).filter(lambda x: x['price'].max()<20000)
print(range2)
```
